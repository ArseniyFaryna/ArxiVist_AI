```python
from pyspark.sql import functions as F

silver_df = spark.table("arxivist_ai_dev.silver.arxiv_cs_papers")

gold_df = silver_df.select(
    F.col("id").alias("chunk_id"),
    F.concat_ws("\n\n", F.col("title"), F.col("abstract")).alias("content"),
    F.col("title"),
    F.col("categories"),
    F.col("authors"),
    F.col("update_date")
).limit(1000)

gold_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("arxivist_ai_dev.gold.arxiv_chunks")

print(f"Gold chunks table created with {spark.table('arxivist_ai_dev.gold.arxiv_chunks').count()} rows.")
display(spark.table("arxivist_ai_dev.gold.arxiv_chunks").limit(5))
```
