```python
# Cleanup: remove index & endpoint 
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()
VS_INDEX_NAME = "arxivist_ai_dev.gold.arxiv_chunks_index"
VS_ENDPOINT_NAME = "arxiv_vs_endpoint"

try:
    w.vector_search_indexes.delete_index(index_name=VS_INDEX_NAME)
    print(f"Index '{VS_INDEX_NAME}' — deleted.")
except Exception as e:
    print(f"Index delete skipped: {e}")

try:
    w.vector_search_endpoints.delete_endpoint(endpoint_name=VS_ENDPOINT_NAME)
    print(f"Endpoint '{VS_ENDPOINT_NAME}' — deleted. No more charges.")
except Exception as e:
    print(f"Endpoint delete skipped: {e}")

import time as _time
_time.sleep(5)
remaining = [ep.name for ep in w.vector_search_endpoints.list_endpoints() if ep.name == VS_ENDPOINT_NAME]
if remaining:
    print(f"\nEndpoint still exists! Delete manually in Compute → Vector Search.")
else:
    print(f"\nCleanup complete. All expensive resources removed.")
```
